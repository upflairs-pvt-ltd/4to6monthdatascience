```python
x =  list(df['day'].value_counts().keys())
y =  df['day'].value_counts().values
print(x)
print(y)
```

```python
plt.bar(x,y)
plt.title('Day wise customer visit')
plt.xlabel('Days name')
plt.ylabel("No. of customers")
plt.show()
```

```python
plt.bar(x,y,color='g')
plt.title('Day wise customer visit')
plt.xlabel('Days name')
plt.ylabel("No. of customers")
plt.show()
```

```python
plt.bar(x,y,color='g',ec='black',lw=4)
plt.title('Day wise customer visit')
plt.xlabel('Days name')
plt.ylabel("No. of customers")
plt.show()
```

```python
plt.bar(x,y,color='g',ec='black',lw=4,ls='--')
plt.title('Day wise customer visit')
plt.xlabel('Days name')
plt.ylabel("No. of customers")
plt.show()
```

```python
color_ls = ['r','g','b','k']
plt.figure(figsize=(12,4))
plt.bar(x,y,ec='k',lw=4,color=color_ls,width=0.6)
plt.title('Day wise customer visit')
plt.xlabel('Days name')
plt.ylabel("No. of customers")
plt.show()
```

```python
color_ls = ['green','white','#FF7722','white']
plt.figure(figsize=(12,7))
plt.barh(x,y,ec='k',lw=4,color=color_ls)
plt.title('Day wise customer visit')
plt.xlabel('Days name')
plt.ylabel("No. of customers")
plt.show()
```
