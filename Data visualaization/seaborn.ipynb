```python
import seaborn as sns  
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# print(sns.load_dataset.__doc__)
```

```python
sns.barplot()
sns.lineplot()
sns.scatterplot()
sns.hist()
sns.displot()
sns.distplot()
sns.countplot()
```

```python
sns.relplot()
sns.catplot()
```

```python
sns.relplot(x='tip',y='total_bill',data=df)
plt.title("total bill Vs Tip")
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,kind='line')
plt.title("total bill Vs Tip")
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,color='r')
plt.title("total bill Vs Tip")
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="sex")
plt.title("total bill Vs Tip")
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="sex",style="smoker")
plt.title("total bill Vs Tip")
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="sex",style="smoker",col="time")

plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="sex",style="smoker",row="time")

plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="sex",style="smoker",col="day")

plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="time",col="day")

plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue="time",col="day",kind='line')

plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='boxen')
```
